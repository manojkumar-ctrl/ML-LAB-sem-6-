# Diabetes Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
print("Columns:", df.columns)
```

```python
X = df.drop("CLASS", axis=1)
y = df["CLASS"]
```

```python
## Handle Missing Values

zero_invalid_cols = ["BMI", "HbA1c", "Urea", "Chol"]
for col in zero_invalid_cols:
    X[col] = X[col].replace(0, np.nan)
```

```python
X_imputed = X.copy() # Initialize X_imputed with a copy of X

# Handle Categorical Data (Gender)
if "Gender" in X_imputed.columns:
    X_imputed = pd.get_dummies(X_imputed, columns=["Gender"], drop_first=True)

#  Handle Outliers (optional but shown here)

for col in X_imputed.select_dtypes(include=[np.number]).columns:
    Q1 = X_imputed[col].quantile(0.25)
    Q3 = X_imputed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    X_imputed[col] = np.clip(X_imputed[col], lower, upper)
```

```python
# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=42)



# MinMax Scaler
mm_scaler = MinMaxScaler()
X_train_mm = mm_scaler.fit_transform(X_train)
X_test_mm = mm_scaler.transform(X_test)

#  Standard Scaler
std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_test_std = std_scaler.transform(X_test)
```

## Adult income data preproceessing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("adult.csv")
```

```python
# Display basic info
print(df.info())
print(df.head())
```

```python
X = df.drop("income", axis=1)
y = df["income"]
```

```python
#  Handle missing values

X = X.replace('?', np.nan)
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns
```

```python
num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")
```

```python
# Preprocess categorical: impute + one-hot encode
numeric_transformer = ColumnTransformer([
    ("num_imputer", num_imputer, num_cols)
], remainder="drop")

categorical_transformer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_cols)
], remainder="drop")
```

```python
# Apply imputers separately
X_num = pd.DataFrame(num_imputer.fit_transform(X[num_cols]), columns=num_cols)
X_cat = pd.DataFrame(cat_imputer.fit_transform(X[cat_cols]), columns=cat_cols)
```

```python
# One-Hot Encode categorical data
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
X_cat_encoded = pd.DataFrame(
    encoder.fit_transform(X_cat),
    columns=encoder.get_feature_names_out(cat_cols)
)
```

```python
X_clean = pd.concat([X_num.reset_index(drop=True), X_cat_encoded.reset_index(drop=True)], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.3, random_state=42)
```

scaling of adult income

```python
# (a) Min-Max Scaler
mm_scaler = MinMaxScaler()
X_train_mm = mm_scaler.fit_transform(X_train)
X_test_mm = mm_scaler.transform(X_test)

# (b) Standard Scaler
std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_test_std = std_scaler.transform(X_test)
```
